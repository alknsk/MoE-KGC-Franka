```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体支持
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]

# 随机生成数据
def generate_random_data():
    """生成随机数据，可能是正态分布、均匀分布或指数分布"""
    distributions = {
        'normal': lambda: np.random.normal(
            loc=np.random.uniform(-10, 10),  # 随机均值
            scale=np.random.uniform(1, 5),   # 随机标准差
            size=np.random.randint(100, 1000)  # 随机数据量
        ),
        'uniform': lambda: np.random.uniform(
            low=np.random.uniform(-20, 0),   # 随机下界
            high=np.random.uniform(0, 20),   # 随机上界
            size=np.random.randint(100, 1000)
        ),
        'exponential': lambda: np.random.exponential(
            scale=np.random.uniform(1, 5),   # 随机比例参数
            size=np.random.randint(100, 1000)
        )
    }
    
    # 随机选择一种分布
    distribution_name = np.random.choice(list(distributions.keys()))
    return distributions[distribution_name](), distribution_name

# 生成随机数据
data, dist_type = generate_random_data()

# 创建直方图
plt.figure(figsize=(10, 6))  # 随机大小
n, bins, patches = plt.hist(
    data,
    bins=np.random.randint(5, 50),  # 随机 bin 数量
    density=np.random.choice([True, False]),  # 随机是否归一化
    color=np.random.rand(3,),  # 随机颜色
    alpha=np.random.uniform(0.5, 1.0)  # 随机透明度
)

# 添加标题和标签
plt.title(f'随机生成的{dist_type}分布直方图')
plt.xlabel('数值')
plt.ylabel('频率')

# 添加网格线
plt.grid(axis='y', alpha=0.75)

# 保存图像（可选）
if np.random.choice([True, False]):
    plt.savefig('random_histogram.png', dpi=300, bbox_inches='tight')

# 显示图形
plt.tight_layout()
plt.show()    
```
