```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]

# 
def generate_random_data():
    """"""
    distributions = {
        'normal': lambda: np.random.normal(
            loc=np.random.uniform(-10, 10),  # 
            scale=np.random.uniform(1, 5),   # 
            size=np.random.randint(100, 1000)  # 
        ),
        'uniform': lambda: np.random.uniform(
            low=np.random.uniform(-20, 0),   # 
            high=np.random.uniform(0, 20),   # 
            size=np.random.randint(100, 1000)
        ),
        'exponential': lambda: np.random.exponential(
            scale=np.random.uniform(1, 5),   # 
            size=np.random.randint(100, 1000)
        )
    }
    
    # 
    distribution_name = np.random.choice(list(distributions.keys()))
    return distributions[distribution_name](), distribution_name

# 
data, dist_type = generate_random_data()

# 
plt.figure(figsize=(10, 6))  # 
n, bins, patches = plt.hist(
    data,
    bins=np.random.randint(5, 50),  #  bin 
    density=np.random.choice([True, False]),  # 
    color=np.random.rand(3,),  # 
    alpha=np.random.uniform(0.5, 1.0)  # 
)

# 
plt.title(f'{dist_type}')
plt.xlabel('')
plt.ylabel('')

# 
plt.grid(axis='y', alpha=0.75)

# 
if np.random.choice([True, False]):
    plt.savefig('random_histogram.png', dpi=300, bbox_inches='tight')

# 
plt.tight_layout()
plt.show()    
```
